```python
import pandas as pd

paths = { "C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/purse_seines.csv","C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/fixed_gear.csv", "C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/trawlers.csv","C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/drifting_longlines.csv","C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/trollers.csv"}

for path in paths:
    df = pd.read_csv(path, header="infer", sep=",",error_bad_lines=False, engine ='python')
    duplicados = df.duplicated(subset='timestamp').sum()
    null_data = df[df.isnull().any(axis=1)].count()
    print(f'{path}\nNulls={null_data}\nDuplicados={duplicados}')
    print("######")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
paths = { "C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/purse_seines.csv","C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/fixed_gear.csv", "C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/trawlers.csv","C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/drifting_longlines.csv","C:/Users/nico_/Desktop/ITBA/TFI/global fishing watch/dataset/trollers.csv"}

for path in paths:
    df = pd.read_csv(path, header="infer", sep=",",error_bad_lines=False, engine ='python')
    df = df.dropna()
    df['timediff']= df.groupby('mmsi')['timestamp'].diff()
    df = df[(df['timediff'] <= 60*30) ]
    df = df[(df['timediff'] > 0) ]
    plt.hist(df['timediff']/60,bins=100)
    plt.ylabel('Frecuencia')
    plt.xlabel('Minutos')
    title = (path.split('/')[-1]).split('.')[0]
    plt.title(f'{title}')
    plt.show()
```
